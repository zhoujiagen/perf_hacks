```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.0, 1.0, 10).astype(np.double)
n = np.linspace(1, 10, 10).astype(np.int)
for _n in n:
    y = np.array([1.0 / (1 - p + p / _n) for p in x])
    plt.plot(x, y, label='n='+str(_n))
plt.legend()
plt.xlabel('p')
plt.ylabel('S')
#plt.rc('text', usetex=True)
#plt.title(r"$S = \frac{1}{1-p+\frac{p}{n}}$") 
plt.title("S = 1 / (1-p+p/n)")
plt.show()
```
